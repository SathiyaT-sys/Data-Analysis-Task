```python
import matplotlib.pyplot as plt
import pandas as pd                                               # Using Pandas and Matplotlib Library
import numpy as np
import seaborn as sns


```

```python
df = pd.read_csv('data.csv')                                         # Load a CSV file
```

```python
print("First 5 rows of the dataset:")
print(df.head())              
print(df.describe())                                                      # Descriptive Statistics
```

```python
column_name="marks"
average_value = df[column_name].mean()                                # Calculating the average
print(f"\n Average of {column_name}:{average_value}")
```

```python
x=np.array(["age","marks","internet","hours"])
y=np.array([15,78.5,1,6.83])
plt.bar(x,y,width=0.3,color=['green','orange'])                                         # Data Visualization using Barchart
plt.title("Bar Chart")
plt.show()
```

```python
plt.scatter(x='hours', y='marks',data=df)                                          # Data Visualization Using Scatter plot
plt.plot()
plt.title("Scatter plot")
plt.xlabel("hours")
plt.ylabel("marks")
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.heatmap(df.corr(),annot=True, linewidth=0.5,cmap='coolwarm')                               # Correlation analysis using Heatmap
plt.title("Heatmap")
```
